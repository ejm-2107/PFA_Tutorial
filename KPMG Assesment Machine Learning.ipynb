```python
import pandas as pd
```

```python
df = pd.read_csv('Hospital.csv')
df.head()
```

```python
# Check missing values
df.isnull().sum()
```

```python
null_index = df.isnull().sum()[df.isnull().sum() > 0].index.values
```

```python
df.info()
```

```python
df.skew()[null_index]
```

```python
# skew between -1 to 1 so mean substitution else median
df['education'].fillna(df['education'].mode()[0],inplace=True)
df['cigsPerDay'].fillna(df['cigsPerDay'].median(),inplace=True)
df['BPMeds'].fillna(df['BPMeds'].median(),inplace=True)
df['totChol'].fillna(df['totChol'].mean(),inplace=True)
df['BMI'].fillna(df['BMI'].mean(),inplace=True)
df['heartRate'].fillna(df['heartRate'].median(),inplace=True)
df['glucose'].fillna(df['glucose'].median(),inplace=True)
```

```python
# Check missing values
df.isnull().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
df_dummy = pd.get_dummies(df,drop_first=True)
df_dummy.shape
```

```python
df_dummy.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_dummy)
scaled_data
```

```python
df_final = pd.DataFrame(scaled_data,columns=df_dummy.columns)
df_final.head()
```

```python
df_final['CHD chance'] = df_dummy['CHD chance']
df_final.head()
```

```python
y = df_final[['CHD chance']] #Output
x = df_final.drop(columns=['CHD chance']) #Input
```

```python
x.head()
```

```python
# Split the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=8748437)
```

```python
len(x_train),len(x_test),len(y_train),len(y_test)
```

## Use logistic regression to identify the accuracy

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(x_train,y_train)
```

```python
y_test['Predicted_LR'] = model.predict(x_test)
y_test
```

```python
# Find the accuracy (classification report)
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test['CHD chance'],y_test['Predicted_LR']))
```

## Use Decision Tree to identify the accuracy

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
model = dt.fit(x_train,y_train)
```

```python
y_test['Predicted_DT'] = model.predict(x_test)
y_test
```

```python
print(classification_report(y_test['CHD chance'],y_test['Predicted_DT']))
```

## Use Random Forest to identify the accuracy

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
model = rf.fit(x_train,y_train)
```

```python
y_test['Predicted_RF'] = model.predict(x_test)
y_test
```

```python
print(classification_report(y_test['CHD chance'],y_test['Predicted_RF']))
```

Accuracy for random forest is 0.84
